* Avg. Session Length: Average session of in-store style advice sessions.

## Training and Testing Data

**Now spliting the data into training and testing sets.
 By setting a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column. **

```python
X=customers[['Avg. Session Length','Time on App','Time on Website','Length of Membership']]
```

```python
y=customers['Yearly Amount Spent']
```

** Using model_selection.train_test_split from sklearn to split the data into training and testing sets.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Training the Model

** Importing LinearRegression from sklearn.linear_model **

```python
from sklearn.linear_model import LinearRegression
```

**Creating an instance of a LinearRegression() model named lm.**

```python
lm=LinearRegression()
```

** Train/fit lm on the training data.**

```python
lm.fit(X_train, y_train)
```

**Printing out the coefficients of the model**

```python
print('coefficients:\n', lm.coef_)
```

## Predicting Test Data
After fitting this model, evaluate its performance by predicting off the test values

** Using lm.predict() to pr dict off the X_test set of the data.**

```python
predictions=lm.predict(X_test)
```

** Creating a scatterplot of the real test values versus the predicted values. **

```python
plt.scatter(y_test,predictions)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

## Evaluating the Model

** Calculating the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

**Ploting a histogram of the residuals and making sure that it looks normally distributed. Using seaborn distplot.**

```python
sns.distplot((y_test-predictions),kde=True,bins=30,hist_kws=dict(edgecolor="k",linewidth=2))
```

## Conclusion

** Recreating the dataframe below. **

```python
coeffecients = pd.DataFrame(lm.coef_,X.columns)
coeffecients.columns = ['Coeffecient']
coeffecients
```

Interpreting the coefficients:

- Holding all other features fixed, a 1 unit increase in **Avg. Session Length** is associated with an **increase of 25.98 total money spent**.
- Holding all other features fixed, a 1 unit increase in **Time on App** is associated with an **increase of 38.59 total money spent**.
- Holding all other features fixed, a 1 unit increase in **Time on Website** is associated with an **increase of 0.19 total money spent**.
- Holding all other features fixed, a 1 unit increase in **Length of Membership** is associated with an **increase of 61.27 total money spent**.
